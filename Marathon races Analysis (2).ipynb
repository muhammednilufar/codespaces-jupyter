```python
#combine all the filters
```

```python
df2=df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event']==2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) =='IND')]
```

```python
df2.shape
```

```python
#remove IND from event name
```

```python
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name']=df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head(10)
```

```python
#cleanup athlete age
```

```python
df2['athlete_age']=2020 - df2['Athlete year of birth']
```

```python
#emove h from athletic peformance
```

```python
df2['Athlete performance']=df2['Athlete performance'].str.split().str.get(0)
```

```python
df2.head(5)
```

```python
#drop columns
```

```python
df2=df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'],axis =1)
```

```python
df2.head()
```

```python
#clean up null values
```

```python
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
df2=df2.dropna()
```

```python
df2.shape
```

```python
#check for dupes
```

```python
df2[df2.duplicated()==True]
```

```python
#reset index
```

```python
df2.reset_index(drop=True,inplace=True)
```

```python
#fix types
```

```python
df2.dtypes
```

```python
df2['athlete_age']=df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed']=df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
#rename columns
```

```python
df2.columns
```

```python
df2=df2.rename(columns = { 'Year of event':'year' ,
               'Event dates': 'race_day',
               'Event name': 'race_name',
               'Event distance/length':'race_length',
               'Event number of finishers':'race_number_of_finishers',
               'Athlete performance':'athlete_performance',
                'Athlete gender':'athlete_gender',
               'Athlete average speed':'athlete_avg_speed',
               'Athlete ID':'athlete_id'})
```

```python
df2.head()
```

```python
#reorder columns
```

```python
df3=df2[['race_day','race_name','race_length','race_number_of_finishers','athlete_id','athlete_gender','athlete_age','athlete_performance','athlete_avg_speed','year']]
```

```python
df3.head()
```

```python
sns.histplot(df3, x= 'race_length', hue='athlete_gender')
plt.title('Number of Racers')
plt.show()
```

```python
sns.displot(df3[df3['race_length']=='50km']['athlete_avg_speed'])
plt.title('Avg Speed of Athletes')
plt.show()
```

```python
sns.violinplot(data=df3,x='race_length',y='athlete_avg_speed',hue='athlete_gender')
plt.title('Avg Speed of Male and Female')
plt.show()
```

```python
sns.lmplot(data=df3,x='athlete_age',y='athlete_avg_speed',hue ='athlete_gender')
plt.title('Performance of Male and Female Racers')
plt.show()
```

```python
#difference in speed for 50km male and female racers
```

```python
df3.groupby(['race_length','athlete_gender'])['athlete_avg_speed'].mean()
```

```python
#what age group are best in the 50km race 
```

```python
df3.query('race_length == "50km"').groupby('athlete_age')['athlete_avg_speed'].agg(['mean','count']).sort_values('mean',ascending=False).query('count>5').head(15)
```

```python
#what age group are worst in the 50km race
```

```python
df3.query('race_length == "50km"').groupby('athlete_age')['athlete_avg_speed'].agg(['mean','count']).sort_values('mean',ascending=True).query('count>5').head(15)
```

```python
#seasons for the data --slower in summer than winter?
#spring 3-5
#summer 6-8
#fall 9-11
#winter 12-2
```

```python
df3['race_month']=df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season']=df3['race_month'].apply(lambda x:'winter' if x>11 else 'fall' if x>8 else 'summer' if x>5 else 'spring' if x>2 else 'winter' )
```
